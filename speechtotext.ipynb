```python
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition", model="openai/whisper-base.en")
```

```python
import gradio as gr
```

```python
# def transcribe(audio):
#   text = p(audio)["text"]
#   return text
```

```python
def transcribe(audio, state=" "):
  """ Speech to text function using the pipeline that we defined """
  text = p(audio)["text"]
  state += text + " "
  return state, state
```

```python
demo = gr.Interface(
    transcribe,
    gr.Audio(sources="microphone"),
    "text",
)

demo.launch()
```
